```python
import pandas as pd
import matplotlib.pyplot as plt

datos = {
    'Comunas': ["Santiago", "Cerrillos", "Cerro Navia", "Conchalí", "El Bosque", "Estación Central", "Huechuraba", "Independencia", "La Cisterna", "La Florida"],
    'Población': [404495, 80832, 132622, 126955, 162505, 147041, 98671, 100281, 90119, 366916]
}

df = pd.DataFrame(datos)

df['Niños'] = round(df['Población'] * 0.2)
df['Jovenes'] = round(df['Población'] * 0.15)
df['Adultos'] = round(df['Población'] * 0.45)
df['Adultos Mayores'] = round(df['Población'] * 0.2)
df['Mujeres'] = round(df['Población'] * 0.51)
df['Hombres'] = round(df['Población'] * 0.49)

#grafico

etiquetas = ['Niños', 'Jóvenes', 'Adultos', 'Adultos Mayores', 'Mujeres', 'Hombres']

valores = [
    df['Niños'].sum(),
    df['Jovenes'].sum(),
    df['Adultos'].sum(),
    df['Adultos Mayores'].sum(),
    df['Mujeres'].sum(),
    df['Hombres'].sum()
]

plt.figure(figsize=(8, 8))
plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', startangle=190)
plt.title('Desglose poblacional')
plt.axis('equal') 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

datos = {
    'Comunas': ["Santiago", "Cerrillos", "Cerro Navia", "Conchalí", "El Bosque", "Estación Central", "Huechuraba", "Independencia", "La Cisterna", "La Florida"],
    'Población': [404495, 80832, 132622, 126955, 162505, 147041, 98671, 100281, 90119, 366916]
}

df = pd.DataFrame(datos)

df['Niños'] = round(df['Población'] * 0.2)
df['Jovenes'] = round(df['Población'] * 0.15)
df['Adultos'] = round(df['Población'] * 0.45)
df['Adultos Mayores'] = round(df['Población'] * 0.2)
df['Mujeres'] = round(df['Población'] * 0.51)
df['Hombres'] = round(df['Población'] * 0.49)

#grafico
plt.figure(figsize=(10, 6))

ancho = 0.2
posiciones = range(len(df['Comunas']))
colores = ['lightblue', 'lightpink', 'lightgrey']

plt.bar([x - ancho for x in posiciones], df['Niños'], width=ancho, color=colores[0], label='Niños')
plt.bar(posiciones, df['Mujeres'], width=ancho, color=colores[1], label='Mujeres')
plt.bar([x + ancho for x in posiciones], df['Hombres'], width=ancho, color=colores[2], label='Hombres')

for i, valor in enumerate(df['Hombres']):
    plt.text(i + ancho, valor - 3000, int(valor), ha='center')
for i, valor in enumerate(df['Mujeres']):
    plt.text(i, valor + 1000, int(valor), ha='center')
for i, valor in enumerate(df['Niños']):
    plt.text(i - ancho, valor - 1000, int(valor), ha='center')

plt.xticks(posiciones, df['Comunas'], rotation=45, ha='right')
plt.xlabel('Comunas')
plt.ylabel('Número de Personas')
plt.title('Desglose por comunas')
plt.legend()

plt.show()
```
